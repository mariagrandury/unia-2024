### Evalúa tu modelo con un script

```python
from lm_eval import evaluator

task = "noticia"

task_result = evaluator.simple_evaluate(
    model="huggingface",  # "hf-causal"
    model_args={"pretrained": "flax-community/gpt-2-spanish"},
    tasks=[task],
    num_fewshot=0,
    batch_size=1,
    device="cuda:0",
    limit=1,
)
task_result["results"][task]
```
